```python
# Read data from CSV file
df = pd.read_csv('labeled_data.csv')
```

```python
df['labels'] = df['class'].map({0: "Hate Speech Detected", 1: "Offensive Language Detected", 2 : "No Hate and Offensive Speech Detected"})
df = df[["tweet", "labels"]]
```

```python
df.head()
```

```python
# preprocessing text
import re
import string
stop_words = set(stopwords.words('english'))
def clean (text):
 text = str(text).lower()
 text = re.sub('\[.*?\]', '', text)
 text = re.sub('https?://\S+|www.\S+', '', text)
 text = re.sub('<.*?>+', '', text)
 text = re.sub(r'[^\w\s]','',text)
 text = re.sub('\n', '', text)
 text = re.sub('\w\d\w', '', text)
 text = [word for word in text.split(' ') if word not in stop_words]
 text=" ".join(text)
 text = [stemmer. stem(word) for word in text. split(' ')]
 text=" ".join(text)
 return text
df["tweet"] = df["tweet"].apply(clean)
```

```python
df.tail()
```

```python
# y = mX + c
x=np.array(df["tweet"]) 
y=np.array(df["labels"])
```

```python
# CountVectorizer is a text preprocessing technique commonly used in natural language processing (NLP) tasks 
# for converting a collection of text documents into a numerical representation
cv = CountVectorizer()
```

```python
x = cv.fit_transform(np.array(df["tweet"])) 
y = np.array(df["labels"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
# Training the model
model = svm.SVC(kernel='linear', C=1.0)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#check Accuracy of The model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
